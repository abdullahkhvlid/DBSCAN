```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_moon = scaler.fit_transform(moon[0])
scaled_moon
```

```python
from sklearn.cluster import DBSCAN
model = DBSCAN(eps=0.3, min_samples=5)
model.fit(scaled_moon)
```

```python
model.get_params()
```

```python
labels = model.labels_
print(labels)
print(len(labels))
```

```python
plt.scatter(scaled_moon[:,0], scaled_moon[:,1], c= labels, s=50, cmap='autumn')
plt.show()
```
